# HW 3: Degeneracy of a particle in a 3D cubic box
The following script computes the degeneracy of each energy level in a particle in a cubic box. The energy of a particle in a 3D cubic box is given by: $$E(n_x,n_y,n_z)=(n_x^2+n_y^2+n_z^2)\times\frac{h^2}{8mL^2},$$ where $$n_x,n_y,n_z$$ are the quantum numbers, $$m$$ denotes the particle mass, and $$L$$ is the box length. For convenience we use "box units", setting $$h=L=m=1.$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def energy_level(n, L):
    """
    Calculate the energy level for a particle in a 3D cubic box
    
    Args:
        n (tuple): tuple of quantum numbers n_x, n_y, n_z
        L (float): length of the cubic box
        
    Returns:
        float: energy level for the given quantum numbers and box length
    """
    nx, ny, nz = n
    return (nx**2 + ny**2 + nz**2)

```

```python
def degeneracy(E, L):
    """
    Calculate the degeneracy as a function of energy for a particle in a 3D cubic box
    
    Args:
        E (float): energy level
        L (float): length of the cubic box
        
    Returns:
        int: degeneracy at the given energy level and box length
    """
    max_n = int(np.sqrt(E))  # maximum quantum number to consider
    count = 0
    for nx in range(1, max_n+1):
        for ny in range(1, max_n+1):
            for nz in range(1, max_n+1):
                if energy_level((nx, ny, nz), L) == E:
                    count += 1
                    print("count:", count)
                    print("nx:",nx, "ny:",ny, "nz:",nz, "E=",energy_level((nx, ny, nz), L))
    return count
```

```python
# create array of energy levels and calculate degeneracy for each level
L = 1  # length of cubic box
#E_array = np.linspace(0, 10, 10)  # array of energy levels
E_array = list(range(1, 60))
print(E_array)
degeneracy_array = np.array([degeneracy(E, L) for E in E_array])

```

```python
# plot degeneracy as a function of energy
plt.plot(E_array, degeneracy_array)
plt.xlabel("Energy (in units of h_bar^2/2mL^2)")
plt.ylabel("Degeneracy")
plt.title("Degeneracy as a Function of Energy\nfor a Particle in a 3D Cubic Box with L = 1")
plt.show()
```
