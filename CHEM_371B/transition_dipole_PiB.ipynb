# HW 7: Transition dipole moment of a particle in a box 

The general expression of the particle in a box wavefunction is given by $\Psi_n(x)=\sqrt{\frac{2}{L}}\sin(\frac{n\pi}{L} x)$. According to quantum mechanics, the intensity of a spectroscopic transition is proportional to the square of the transition dipole moment: $\mu_{if}=\int \Psi_i \hat{x} \Psi_f dx, $
where $\Psi_i$ and $\Psi_f$ denote initial and final state, respectively.

1. Use the following python script to compute the transition dipole moments between the ground state ($n_i=1$) and higher excited states ($n_f = 2, 3, 4, 5$). To do so modify the final state quantum number.

2. Verify the symmetry-based selection rules by inspecting the symmetry of initial and final wavefunctions using the plot.

Extra credit: modify the script such it computes the transition dipole moments simoultaneaously for several states and plot the stick absorption spectrum of the particle in a box.  


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.01)  # x-axis

L = 20
n_i = 1    #initial state quantum number


n_f = 3    #final state quantum number

psi_i = np.sqrt(2 / L) * np.sin(n_i * np.pi / L * (x + 10))

psi_f = np.sqrt(2 / L) * np.sin(n_f * np.pi / L * (x + 10))

new_function = psi_i * x * psi_f
transdip = np.sum(new_function) * 0.01
```

```python
plt.figure()
plt.plot(x, psi_i, label='initial WF')
plt.plot(x, psi_f, label='final WF')
plt.plot(x, new_function, label='initial_WF * x * final_WF')
plt.legend()
plt.show()

```

```python
print(transdip)
```

```python
n_f= [2, 3, 4, 5, 6]
trans_dip = []  # List to store transition dipole moments (y-axis of the stick spectrum)
energy_gap = [] # List of excitation energies levels (x-axis of the stick spectrum)
```

```python
for n in n_f:
    psi_f = np.sqrt(2 / L) * np.sin(n * np.pi / L * (x + 10))

    new_function = psi_i * x * psi_f
    dipole_moment = np.sum(new_function) * 0.01
    trans_dip.append(dipole_moment**2)
    energy_gap.append(2 * n -1)
    
```

```python
print(trans_dip)
```

```python
print(energy_gap)
```

```python
# Create a bar plot (stick plot)
plt.bar(energy_gap, trans_dip)

# Add labels and title
plt.xlabel('Energy / Box units')
plt.ylabel('Transition dipole moment')
plt.title('Stick Spectrum')

# Show the plot
plt.show()
```
