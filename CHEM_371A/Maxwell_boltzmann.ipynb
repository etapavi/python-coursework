```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k_B = 1.38e-23  # Boltzmann constant (J/K)
m = 1.0e-27     # Particle mass (kg)

# Function to calculate the Maxwell-Boltzmann distribution
def maxwell_boltzmann(speed, T):
    return 4 * np.pi * (speed**2) * (m / (2 * np.pi * k_B * T))**(3/2) * np.exp(-(m * (speed**2)) / (2 * k_B * T))

# Temperature range and speed values
temperatures = [300, 500, 1000]  # Temperatures in Kelvin
speeds = np.linspace(0, 20000, 1000)  # Speed values from 0 to 2000 m/s

# Plot Maxwell-Boltzmann distributions for each temperature
for T in temperatures:
    distribution = maxwell_boltzmann(speeds, T)
    norm=np.sum(distribution)
    #print(norm)
    distribution=distribution/norm
    plt.plot(speeds, distribution, label=f'T = {T} K')
    

# Plot settings
plt.title('Maxwell-Boltzmann Distribution')
plt.xlabel('Speed (m/s)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
